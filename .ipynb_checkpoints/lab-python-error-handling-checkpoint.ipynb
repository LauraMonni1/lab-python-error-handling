def initialize_inventory(products):

                quantity = int(input(f"Enter the quantity of {product}s available: "))

````python
def initialize_inventory (products:list):
    
    """the function initializes the inventory and can handle typing errors from the user.
    The function returns the dictionary 'inventory' and prints the products and
    relative quantities in separate lines"""
    
    inventory = {}  # create an empty dictionary called inventory
    
    for product in products: # loop to add the number of products and create the inventory dictionary
       
        valid_user_input = False
        while not valid_user_input:
            try:
                user_input = int(input(f"How many {product}s are there? "))
                if user_input >= 0:
                    inventory[product] = user_input
                    valid_user_input = True
                else:
                    print("Invalid! The product's quantity cannot be negative.")
               
            except ValueError: 
                print("Error! Please enter a valid number.")           
    
    print() # space line
    print("The inventory contains: ")
    for product, quantity in inventory.items(): # loop to print the dictionary keys and values in separate lines so the user knows how many products are there 
        print(f"{product}s: {quantity}")
    
    return inventory 
````

````python
def get_customer_orders(inventory):
    
    """the function prompts the user to enter the number of product the customer wants to order.
    Then it asks the name of the products the customer wants and repeat the question for the number of orders.
    The functions handles ValueErrors, messages if the product is not in the inventory, 
    if the product was already ordered or if the product is out of stock.
    The customer orders are then stored in a set called 'customer_orders' that is finally returned by the function."""
    
    customer_orders=set()
    
    while True:
        try:
            number_orders = int(input("Please enter the number of customer orders: "))
            break
        except ValueError:
            print("Error! Please enter a valid number.")
            
    
    for customer_input in range(number_orders):
          
        while True:  # Start of the while loop that allows the user to add ordered products until needed
            try:
                customer_input = input("Enter a product the customer wants to order: ").lower()
                if customer_input in inventory: # checks if the entered product is actually in the inventory
                    if inventory[customer_input] > 0: # checks if the product is not out of stock
                        if customer_input not in customer_orders: # checks if the product was already ordered
                            customer_orders.add(customer_input)
                            print(f"{customer_input} was added to the order.")
                            break
                            
                        else:
                            print("This product is already in the shopping basket. Please order a new one.")
                    
                    elif inventory[customer_input] == 0:
                        print("Sorry, this product is out of stock.")
                        
                else:
                    print("Sorry, we don't have this product.") 
                        
            except:
                print("Error! Please enter a valid name.")
                
                
                               
    return customer_orders   
````

````python
def update_inventory(customer_orders: set, inventory: dict):
    
    updated_inventory = {product: (inventory[product]- 1 if product in customer_orders else inventory[product]) for product in inventory.keys()} # dict comprehension
    updated_inventory = {product: updated_inventory[product] for product in updated_inventory.keys() if updated_inventory[product] > 0} 
    
    return updated_inventory
````

````python
def calculate_order_statistics(customer_orders:set, products:list):
    
    """This function calculates the statistics of the order 
    returning a tuple containing the total number of products ordered and the 
    percentange of the ordered product in the inventory"""
    
    products = list(inventory.values())
    
    total_order = len(customer_orders)
    total_inventory = sum(products)
    percentage_order = ((total_order/total_inventory)*100)  # % of ordered products in the inventory
    order_status = (total_order, percentage_order) # tuple storing the statistics of ordered products
    
    return order_status  
````

````python
def print_order_statistics(order_status):
    
    """The function allows to print the order_status tuple with the statistics of the order"""
    
    total = order_status[0]
    percentage = order_status[1]
    
    print()  # space line
    print(f"""Order Statistics: 
    Total Products Ordered:{total}
    Percentage of Products Ordered: {percentage: .1f}%""")
    
    return
````

````python
def print_updated_inventory(updated_inventory:dict):
    
    print() # space line
    
    print("The updated inventory contains:")
    {print(f"{product}s: {quantity}") for product, quantity in updated_inventory.items()} # dict comprehension
    
    print() # space line    
    
    return
````

````python
def print_price(customer_orders):
    
    product_price = {}
    
    for product in customer_orders:
        
        while True:
            try:
                price_input = float(input(f"Enter the price of {product:}: ")) 
                if price_input >= 0:
                    product_price[product] = price_input
                    break
                else:
                    print("Error! Please enter a valid price.")
            
            except ValueError:
                print("Error! Please enter a valid price.")
                   
    print(f"Total price: {sum(product_price.values()): .1f} €")
    
    return
````

````python
inventory = initialize_inventory(["t-shirt", "mug", "hat", "book", "keychain"])

customer_orders = get_customer_orders(inventory)

updated_inventory = update_inventory(customer_orders, inventory)

my_stats = calculate_order_statistics(customer_orders, inventory)

print_order_statistics(my_stats)

print_updated_inventory(updated_inventory)

print_price(customer_orders)
````
